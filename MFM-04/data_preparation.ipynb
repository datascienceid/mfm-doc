## **Handling Missing & Duplicated Values**

```python
#check empty values
df.isna().sum()
```

```python
#Drop all empty values
df_remove_all = df.dropna()
df_remove_all
```

```python
#drop selected columns
df_remove_selected = df.dropna(subset= ['Description'])
df_remove_selected
```

```python
#handling empty values: numerical imputation
df['CustomerID'].fillna(0).reset_index()
```

```python
#handling empty values: numerical imputation
df['CustomerID'].fillna(df['CustomerID'].mean()).reset_index()
```

```python
#handling empty values: categorical imputation
df['Description'].fillna('Description')
```

```python
#fill with mode
df['Description'].fillna(df['Description'].mode()[0])
```

```python
#Handling Duplicate Values

'''
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop_duplicates.html
'''

df.duplicated().sum()
```

```python
df.drop_duplicates(keep = False)
```

```python
#handling outliers: box plot
import matplotlib.pyplot as plt

# Set the figure size
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

data = pd.DataFrame({"Quantity": df['Quantity']})

# Plot the dataframe
ax = data[['Quantity']].plot(kind='box', title='Find Outliers')
plt.show()
```

```python
#handling outliers: IQR
Q1 = df['Quantity'].quantile(0.25)
Q3 = df['Quantity'].quantile(0.75)
IQR = Q3 - Q1

low_limit = Q1 - (1.5 * IQR)
high_limit = Q3 + (1.5 * IQR)

df_filtered = ((df['Quantity'] >= low_limit) & (df['Quantity'] <= high_limit))
```

```python
df_new = df[df_filtered]
df_new
```

```python
#Handling Outliers: Z-Score

z_score = np.abs(stats.zscore(df['Quantity']))
z_score
```

```python
filtered = z_score > 3
df_z = df[filtered]
df_z
```

```python
df.drop(df_z.index)
```

## **Data Transformation**

```python
#Standardization vs Normalization
from sklearn.preprocessing import MinMaxScaler, StandardScaler

df['norm'] = MinMaxScaler().fit_transform(df['UnitPrice'].values.reshape(len(df), 1))
df['standard'] = StandardScaler().fit_transform(df['UnitPrice'].values.reshape(len(df), 1))
```
